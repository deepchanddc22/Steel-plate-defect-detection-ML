```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv(r"D:\data science\Data scientist\Competitions\train.csv")
dft = pd.read_csv(r"D:\data science\Data scientist\Competitions\test.csv")
```

```python
test_feat = dft.drop('id',axis = 1)
```

```python
test_ids = dft['id']
```

```python
X = df.drop(['id', 'Pastry', 'Z_Scratch', 'K_Scatch', 'Stains',
                           'Dirtiness', 'Bumps', 'Other_Faults'], axis=1)
y = df[['Pastry', 'Z_Scratch', 'K_Scatch', 'Stains',
                      'Dirtiness', 'Bumps', 'Other_Faults']]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(n_estimators=1000,
    criterion='gini',
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    min_weight_fraction_leaf=0.0,
    max_features='sqrt',
    bootstrap=True,
    )
rf_multi_classifier = MultiOutputClassifier(rf_classifier,n_jobs = -1)

```

```python
rf_multi_classifier.fit(X_train,y_train)
```

```python
y_pred = rf_multi_classifier.predict(X_test)

from sklearn.metrics import classification_report,accuracy_score

accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
best_test_probs = rf_multi_classifier.fit(X_train,y_train).predict_proba(test_feat)
best_test_probs
```

```python
import time
submission_df = pd.DataFrame({'id': test_ids})
for i, target in enumerate(y_train.columns):
    submission_df[target] = best_test_probs[i][:, 1]  
    time.sleep(6)
    print(submission_df) 
submission_df.to_csv('D:\data science\Data scientist\Competitions\Submission.csv', index=False)
```
